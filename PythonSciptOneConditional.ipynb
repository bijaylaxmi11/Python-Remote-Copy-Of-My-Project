5. Check whether a number is positive, negative, or zero.

```python
try :
     n = float(input("Enter a number : "))
     if(n > 0):
      print(f" {n} is positive")
     elif(n < 0):
      print(f"{n} is negative")
     else:
      print(f"{n} is zero")
except ValueError:
   # Handle the error if the user enters text instead of a number
   print("Error: Invalid input. Please enter a numerical value.")     
```

8. Check if a number is prime.

```python
n = int(input("Enter a number: "))

if n <= 1:
    print("Not a prime number")
else:
    is_prime = True
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime number")
    else:
        print("Not a prime number")


```

9. Check if a year is a leap year.

```python
n = int(input("Enter a number"))
if n % 400 == 0 or (n%4==0 and n%100 !=0 ):
    print("Leap year")
else:
    print("Not a Leap Year")
```

10.The sum of the first n natural numbers

```python
n = int(input("Enter a number (N): "))
total_sum = 0
for i in range(1, n + 1):
    total_sum += i  # Adds current number 'i' to total_sum
print(f"The sum of the first {n} natural numbers is: {total_sum}")
```

11. Multiplication Table of number

```python
# Get input from the user
num = int(input("Enter a number to see its multiplication table: "))

print(f"\n--- Multiplication Table of {num} ---")

# Loop from 1 to 10 (inclusive)
for i in range(1, 11):
  # Print each line using an f-string for formatting
  print(f"{num} x {i} = {num * i}")
```
