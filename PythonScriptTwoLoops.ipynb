1. Print all even numbers between 1 and 50.

```python
for i in range(1,51):
    if i%2 == 0:
        print(i,end = " ")
```

2. Find the factorial of a number.

```python
a = 1
n = int(input("Enter a number : "))
for i in range(1,n+1) :
    a = a * i
print(f"Factorial of {n} is {a}")
```

3. Count digits in a number.

4. Reverse a number.

```python
n = input("Enter a number")
print(n[::-1])
```
