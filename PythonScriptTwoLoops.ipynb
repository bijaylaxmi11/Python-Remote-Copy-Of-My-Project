1. Print all even numbers between 1 and 50.

```python
for i in range(1,51):
    if i%2 == 0:
        print(i,end = " ")
```

2. Find the factorial of a number.

```python
a = 1
n = int(input("Enter a number : "))
for i in range(1,n+1) :
    a = a * i
print(f"Factorial of {n} is {a}")
```

5. Check whether a number is pallindromme

```python
n = input("Enter a number")
if str(n) == str(n)[::-1]:
    print(f"{n} is pallindrome")
else :
    print("not pallindromme")
```

6. Write fibonacci series of n number

```python
n = int(input("Enter a number"))
a,b = 0,1
for _ in range(n):
    print(a,end = " ")
    a,b=b,a+b  
```
